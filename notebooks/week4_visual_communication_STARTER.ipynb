## TUESDAY -- "The Confusion Matrix for Clinicians"


A confusion matrix is a 2×2 (or N×N) table of prediction outcomes. For a radiologist, "False Negative" means a disease was present but the model said it was not. For pneumonia detection, a false negative means a patient is sent home with untreated pneumonia. For breast cancer, a false negative means a malignant tumour is called benign. The clinical consequence of each cell determines the acceptable operating threshold.

## Common Mistakes

**Reporting aggregate metrics only:** "87% weighted F1" tells Nurse Folake nothing. Per-class recall per clinical subtype is the honest metric.

**Methodology-first presentations:** Nurse Folake stops listening after slide 3 if the finding has not appeared. Lead with what she needs to decide.

**Treating false positives and false negatives as equally bad:** In cancer detection, a false negative (missed cancer) is far more costly than a false positive (unnecessary follow-up). The model's threshold is a clinical decision, not a ML default.
